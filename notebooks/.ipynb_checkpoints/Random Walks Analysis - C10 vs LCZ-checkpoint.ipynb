# Analysing Route 1 on LCZ42

```python
import glob2
path_to_eval_file_rw1 = "/media/trao_ka/SAMSUNG/DLR_DAAD_Research_Backup/lfa-rwa-24-11-2020/rwa1_half1/lcz42_sen2_d2_trial_0_incremental_meta.data"
with open(path_to_eval_file_rw1, 'rb') as filehandle:
    # read the data as bina|ry data stream
    half1_rw1_lcz_evals = pkl.load(filehandle)
```

```python
half1_route1_lcz42 = []
for (mat, op) in encodings_r1:
    for (mat_, op_, dt) in half1_rw1_lcz_evals['trial_0']:
        
        mat = np.array(mat)
        cell = api.ModelSpec(
         matrix=mat.astype(int),
         # Operations at the vertices of the module, matches order of matrix.
         ops=op)
        hash_0 = cell.hash_spec(op)
    
        mat_ = np.array(mat_)
        cell_prime = api.ModelSpec(
             matrix=mat_.astype(int),
             # Operations at the vertices of the module, matches order of matrix.
             ops=op_)       
        #print(hash_0, hash_0_prime)
        hash_0_prime = cell_prime.hash_spec(op_)
        #assert(hash_0 == hash_0_prime)
        
        if hash_0 == hash_0_prime:#np.all(mat == mat_) and np.all(op == op_):
            half1_route1_lcz42.append(dt)
```

```python
path_to_eval_file_rw2 = "/media/trao_ka/SAMSUNG/DLR_DAAD_Research_Backup/lfa-rwa-24-11-2020/"
path_to_eval_file_rw2 += "rwa1_half2/*_meta.data"

more_dfiles_2 = glob2.glob(path_to_eval_file_rw2)

half2_route1_lcz42 = []

cpt = 0
for (mat, op) in encodings_r1[50:]:
    mat = np.array(mat)
    cell = api.ModelSpec(
         matrix=mat.astype(int),
         # Operations at the vertices of the module, matches order of matrix.
         ops=op)
    hash_i = nasbench._hash_spec(cell)
    d = None
    for fname in more_dfiles_2:
        if hash_i in fname:            
            with open(fname, 'rb') as filehandle:
            #    # read the data as binary data stream
                d = pkl.load(filehandle)
                half2_route1_lcz42.append(d)
    if d is None:
        inter_fake_dict = {'training_time': 3600, 'kappaCohen_test':0.99}
        d = {4: inter_fake_dict, 12: inter_fake_dict, 36: inter_fake_dict}
        half2_route1_lcz42.append(d)
            
```

```python
# Smoothed

def movingaverage (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma

```

```python
len_win = 5
y_test_kappa_36_r1_ma = movingaverage(y_test_kappa_36_r1, len_win)
perfs_r1_ma = movingaverage(perfs_r1, len_win)
len_to_remove_lcz = len(all_data_r1) - len(y_test_kappa_36_r1_ma)
```

```python
dict_routes_perfs = dict()
n_routes = 30
n_steps_per_route = 100

for rt_i in range(n_routes):
    route_ri = fitness_walk.loc[fitness_walk['route'] == rt_i + 1]
    dict_routes_perfs[rt_i] = [nasbench_res[int(idx_ri - 1)]['fitness']['36']['validation_acc'] for idx_ri in route_ri['path']]
    assert len( dict_routes_perfs[rt_i]) == n_steps_per_route
```

```python
N_lags_ACF = 1 #len(perfs_r1)/20
```

```python
# Autocorelation with step - 1, for Routes 1 & 16
rho_r1_c10 = acf(np.array(perfs_r1), nlags=N_lags_ACF)[1:].mean()
rho_r16_c10 = acf(np.array(perfs_r16), nlags=N_lags_ACF)[1:].mean()

# Autocorellation length  Tau = 1 / Rho(1)
tau_acf_len_r1_c10  = 1 / rho_r1_c10
tau_acf_len_r16_c10  = 1 / rho_r16_c10

# The Smaller the TAU, the rougher the surface
# Plus, R16 is visually rougher than R1
# So, TAU_R16 should be smaller than TAU_R1
#assert tau_acf_len_r16_c10 < tau_acf_len_r1_c10
```

### Same for all routes in C10

```python
rho_all_routes_c10 = list()
tau_all_routes_c10 = list()
N_lags_ACF = 1

for rt_i in range(n_routes):
    
    # Autocorelation with step - 1, for Routes 1 & 16
    rho_ri = acf(np.array(dict_routes_perfs[rt_i]), nlags=N_lags_ACF)[1:]
    rho_all_routes_c10.append(rho_ri)
    
    
    # Autocorellation length  Tau = 1 / Rho(1)
    tau_acf_len_ri_c10  = 1 / rho_ri
    
    tau_all_routes_c10.append(tau_acf_len_ri_c10)
```

```python
assert len(rho_all_routes_c10) == n_routes
assert len(tau_all_routes_c10) == n_routes
```
