## Across training regimes

```python
len_t0_half = 50

mean_perf_half1 = []
std_perf_half1 = []

for m_i in range(len_t0_half):
    m0, op0, d0 = read_data_1['trial_0'][m_i]
    m1, op1, d1 = read_data_1['trial_1'][m_i]
    m2, op2, d2 = read_data_1['trial_2'][m_i]
    
    assert np.all(m0 == m1) and np.all(m1 == m2)
    assert np.all(op0 == op1) and np.all(op1 == op2)
    
    d_mean = {}
    d_std = {}
    all_regimes = [4, 12, 36, 108]
    d0_keys = [*d0]
    d1_keys = [*d1]
    d2_keys = [*d2]
    assert d0_keys == all_regimes and d1_keys == all_regimes and d2_keys == all_regimes
    
    all_attributes = [*d0[all_regimes[0]]]

    for r_i in all_regimes:
        d_mean[r_i] = {}
        d_std[r_i] = {}
    
        for at_i in all_attributes:
            d_mean[r_i][at_i] = np.mean([d0[r_i][at_i], d1[r_i][at_i], d2[r_i][at_i]])            
            d_std[r_i][at_i] = np.std([d0[r_i][at_i], d1[r_i][at_i], d2[r_i][at_i]])
            #print(d0[r_i][at_i], d0[r_i][at_i], d0[r_i][at_i])
    mean_perf_half1.append((m0, op0, d_mean))
    #std_perf_half1.append(d_std)
```

```python
len_t0_half = 50

mean_perf_half2 = []
std_perf_half2 = []

for m_i in range(len_t0_half):
    m0, op0, d0 = read_data_2['trial_0'][m_i]
    m1, op1, d1 = read_data_2['trial_1'][m_i]
    #m2, op2, d2 = read_data_2['trial_2'][m_i]
    
    assert np.all(m0 == m1) #and np.all(m1 == m2)
    assert np.all(op0 == op1) #and np.all(op1 == op2)
    
    d_mean = {}
    d_std = {}
    all_regimes = [4, 12, 36, 108]
    d0_keys = [*d0]
    d1_keys = [*d1]
    #d2_keys = [*d2]
    assert d0_keys == all_regimes and d1_keys == all_regimes #and d2_keys == all_regimes
    
    all_attributes = [*d0[all_regimes[0]]]

    for r_i in all_regimes:
        d_mean[r_i] = {}
        d_std[r_i] = {}
    
        for at_i in all_attributes:
            d_mean[r_i][at_i] = np.mean([d0[r_i][at_i], d1[r_i][at_i]]) #, d2[r_i][at_i]])            
            d_std[r_i][at_i] = np.std([d0[r_i][at_i], d0[r_i][at_i]]) #, d0[r_i][at_i]])
            
    mean_perf_half2.append((m0, op0, d_mean))
    
    #mean_perf_half2.append(d_mean)
    #std_perf_half2.append(d_std)
```

```python
mean_perf_half = mean_perf_half1 + mean_perf_half2
```

```python
def top_n(list_models, N=10, regime=108, key='kappaCohen_test'):
    
    sorted_by_key = sorted(list_models, key= lambda triplet: triplet[2][regime][key])
    
    top_n_idx = len(sorted_by_key) * N/100
    top_n_idx = round(top_n_idx)
    
    return sorted_by_key[-top_n_idx:]
```

```python
N = 100
top_25_r_04 = top_n(mean_perf_half, N=N, regime=4, key='kappaCohen_test')
top_25_r_12 = top_n(mean_perf_half, N=N, regime=12, key='kappaCohen_test')
top_25_r_36 = top_n(mean_perf_half, N=N, regime=36, key='kappaCohen_test')
top_25_r_108 = top_n(mean_perf_half, N=N, regime=108, key='kappaCohen_test')
```

```python
top_25_r_4_models = [(m4, o4) for (m4, o4, _) in top_25_r_04]
top_25_r_12_models = [(m12, o12) for (m12, o12, _) in top_25_r_12]
top_25_r_36_models = [(m36, o36) for (m36, o36, _) in top_25_r_36]
top_25_r_108_models = [(m108, o108) for (m108, o108, _) in top_25_r_108]
```

```python
cross_4_12 = []
cross_4_12_36 = []
cross_all = []

for duo_i in top_25_r_04:
    # in R = 12 
    for duo_j in top_25_r_12:
        if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
            cross_4_12.append(duo_i)

for duo_i in cross_4_12:
    # in R = 12 
    for duo_j in top_25_r_36_models:
        if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
            cross_4_12_36.append(duo_i)
            
for duo_i in cross_4_12_36:
    # in R = 12 
    for duo_j in top_25_r_108:
        if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
            cross_all.append(duo_i)
```

## Plot evolution of Top-N intersection

```python
def Evol_top_N_relatively_4(relative=False):
    size_inter = []
    total_size = len(mean_perf_half)
    for N in range(1, total_size + 1):
        top_25_r_04 = top_n(mean_perf_half, N=N, regime=4, key='kappaCohen_test')
        top_25_r_12 = top_n(mean_perf_half, N=N, regime=12, key='kappaCohen_test')
        top_25_r_36 = top_n(mean_perf_half, N=N, regime=36, key='kappaCohen_test')
        top_25_r_108 = top_n(mean_perf_half, N=N, regime=108, key='kappaCohen_test')

        top_25_r_4_models = [(m4, o4) for (m4, o4, _) in top_25_r_04]
        top_25_r_12_models = [(m12, o12) for (m12, o12, _) in top_25_r_12]
        top_25_r_36_models = [(m36, o36) for (m36, o36, _) in top_25_r_36]
        top_25_r_108_models = [(m108, o108) for (m108, o108, _) in top_25_r_108]

        cross_4_12 = []
        cross_4_12_36 = []
        cross_all = []

        for duo_i in top_25_r_04:
            # in R = 12 
            for duo_j in top_25_r_12:
                if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
                    cross_4_12.append(duo_i)

        for duo_i in cross_4_12:
            # in R = 12 
            for duo_j in top_25_r_36_models:
                if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
                    cross_4_12_36.append(duo_i)

        for duo_i in cross_4_12_36:
            # in R = 12 
            for duo_j in top_25_r_108:
                if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
                    cross_all.append(duo_i)
        if relative:
            len_top_4 = len(top_25_r_4_models)
            assert len_top_4 == N
        else:
            len_top_4 = total_size
        
        l0, l1, l2, l3 = len(top_25_r_4_models) / len_top_4, len(cross_4_12)/len_top_4, len(cross_4_12_36)/len_top_4, len(cross_all)/len_top_4
        size_inter.append((l0 * 100, l1 * 100, l2 * 100, l3 * 100))
    return size_inter
    
```

## Evolution of Failures: the Bottom-N% performers accross regimes

```python
def bottom_n(list_models, N=10, regime=108, key='kappaCohen_test'):
    
    sorted_by_key = sorted(list_models, key= lambda triplet: triplet[2][regime][key])
    
    bottom_n_idx = len(sorted_by_key) * N/100
    bottom_n_idx = round(bottom_n_idx)
    
    return sorted_by_key[:bottom_n_idx]
```

```python
def Evol_bottom_N_relatively_4(relative=False):
    size_inter = []
    total_size = len(mean_perf_half)
    for N in range(1, total_size + 1):
        bottom_25_r_04 = bottom_n(mean_perf_half, N=N, regime=4, key='kappaCohen_test')
        bottom_25_r_12 = bottom_n(mean_perf_half, N=N, regime=12, key='kappaCohen_test')
        bottom_25_r_36 = bottom_n(mean_perf_half, N=N, regime=36, key='kappaCohen_test')
        bottom_25_r_108 = bottom_n(mean_perf_half, N=N, regime=108, key='kappaCohen_test')

        bottom_25_r_4_models = [(m4, o4) for (m4, o4, _) in bottom_25_r_04]
        bottom_25_r_12_models = [(m12, o12) for (m12, o12, _) in bottom_25_r_12]
        bottom_25_r_36_models = [(m36, o36) for (m36, o36, _) in bottom_25_r_36]
        bottom_25_r_108_models = [(m108, o108) for (m108, o108, _) in bottom_25_r_108]

        cross_4_12 = []
        cross_4_12_36 = []
        cross_all = []

        for duo_i in bottom_25_r_04:
            # in R = 12 
            for duo_j in bottom_25_r_12:
                if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
                    cross_4_12.append(duo_i)

        for duo_i in cross_4_12:
            # in R = 12 
            for duo_j in bottom_25_r_36_models:
                if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
                    cross_4_12_36.append(duo_i)

        for duo_i in cross_4_12_36:
            # in R = 12 
            for duo_j in bottom_25_r_108:
                if np.all(duo_i[0] == duo_j[0]) and np.all(duo_i[1] == duo_j[1]) :
                    cross_all.append(duo_i)
        if relative:
            len_bottom_4 = len(bottom_25_r_4_models)
            assert len_bottom_4 == N
        else:
            len_bottom_4 = total_size
        
        l0, l1, l2, l3 = len(bottom_25_r_4_models) / len_bottom_4, len(cross_4_12)/len_bottom_4, len(cross_4_12_36)/len_bottom_4, len(cross_all)/len_bottom_4
        size_inter.append((l0 * 100, l1 * 100, l2 * 100, l3 * 100))
    return size_inter
```

```python

# Useful constants
INPUT = 'input'
OUTPUT = 'output'
CONV3X3 = 'conv3x3-bn-relu'
CONV1X1 = 'conv1x1-bn-relu'
MAXPOOL3X3 = 'maxpool3x3'
NUM_VERTICES = 7
MAX_EDGES = 9
EDGE_SPOTS = NUM_VERTICES * (NUM_VERTICES - 1) / 2   # Upper triangular matrix
OP_SPOTS = NUM_VERTICES - 2   # Input/output vertices are fixed
ALLOWED_OPS = [CONV3X3, CONV1X1, MAXPOOL3X3]
ALLOWED_EDGES = [0, 1]   # Binary adjacency matrix
```

```python
def sample_single_op(NUM_VERTICES, ALLOWED_OPS):
    x = lhs(NUM_VERTICES, samples=1)[0]
    vv = np.floor(x * len(ALLOWED_OPS))
    op  = [ALLOWED_OPS[int(k)] for k in vv]
    op[0] = INPUT
    op[-1] = OUTPUT
    return op

def sample_single_configurations_lhs(N_dimensions):    
    
    sum_edges = 0   
    while sum_edges != 9:
        v_m = lhs(N_dimensions, samples=1)
        idx = v_m > 0.5
        v_m[idx == True] = 1
        v_m[idx == False] = 0
        sum_edges = sum(v_m[0])
    
    return v_m[0]

def recover_incidence_matrix(a0, N_l=7):
    
    mat = np.zeros((N_l, N_l))
    idx_new = 0
    idx_old = 0
    
    for i in range(N_l):
        idx_new += N_l - i - 1
        
        values = a0[idx_old:idx_new]
        idx_old = idx_new
        mat[i, i+1: ] = values
        
    return mat

def sample_single_valid_spec(NUM_vert, allowed_ops):
    
    is_valid = False
    while not is_valid:
        current_op = sample_single_op(NUM_vert, allowed_ops)
        current_config = sample_single_configurations_lhs(N_dimensions = 7 * 3) 
        current_mat = recover_incidence_matrix(current_config, N_l=NUM_vert)
        current_spec = api.ModelSpec(matrix=current_mat, ops=current_op)        
        is_valid = nasbench.is_valid(current_spec)
        
    return current_mat, current_op, current_spec

def LHS_sample_N_valid_specs(N, NUM_vert, allowed_ops, nb):
 all_specs = []
 set_mats = set()
 while len(set_mats) < N:
   s = sample_single_valid_spec(NUM_vert, allowed_ops)
   m, o, sp = s
   t_m = tuple(m.reshape(NUM_vert * NUM_vert))
   hash_m = nb._hash_spec(sp)
   if hash_m not in set_mats:
     set_mats = set_mats | set([t_m])
     all_specs.append(s)

 return all_specs
```

```python
def random_spec():
  """Returns a random valid spec."""
  while True:
    matrix = np.random.choice(ALLOWED_EDGES, size=(NUM_VERTICES, NUM_VERTICES))
    matrix = np.triu(matrix, 1)
    ops = np.random.choice(ALLOWED_OPS, size=(NUM_VERTICES)).tolist()
    ops[0] = INPUT
    ops[-1] = OUTPUT
    spec = api.ModelSpec(matrix=matrix, ops=ops)
    if nasbench.is_valid(spec):
      return matrix, ops, spec
```

```python
def top_n_local(values_, N=100, key='test_accuracy'):
    s_key = sorted(values_, key= lambda triplet: triplet[2][key]) 
    top_n_idx = len(s_key) * N/100
    top_n_idx = round(top_n_idx)
    return s_key[-top_n_idx:]
```

```python
def bottom_n_local(values_, N=100, key='test_accuracy'):
    s_key = sorted(values_, key= lambda triplet: triplet[2][key]) 
    top_n_idx = len(s_key) * N/100
    top_n_idx = round(top_n_idx)
    return s_key[:top_n_idx]
```
